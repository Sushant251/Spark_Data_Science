## Library imported

```python
import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
data = pd.read_csv('SampleSuperstore.csv')
data.head(10)
```

## Exploring data

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(keep='first',inplace=True)
new_data=data.drop(['Country','Postal Code','Profit','Sales','Quantity','Discount'],axis=1)
new_data
```

## Data Visualisation

```python
col=list(new_data.columns)
for i in col:
    plt.figure(figsize=(12,8))
    sns.countplot(data[i],order=(data[i].value_counts().head(30)).index)
    plt.xticks(rotation=90)
    plt.title(i)
    plt.show()   
```

```python
col=list(new_data.columns)
for i in col:
    
    if(i=='City' or i=='State'):
        continue
    else:
        plt.figure(figsize=(12,8))
        data[i].value_counts().plot.pie(autopct='%1.1f%%')

        plt.show()
```

```python
col=list(new_data.columns)
for i in col:
    
    if(i=='City' or i=='State'):
        continue
    else:
        plt.figure(figsize=(12,8))
        data.groupby(i)['Profit','Sales'].agg(['sum']).plot.bar()

        plt.show()

data.groupby('Discount')['Profit','Sales'].agg(['sum']).plot.bar()
data.groupby('Quantity')['Profit','Sales'].agg(['sum']).plot.bar()        
```

```python
plt.figure(figsize=(15,9))
sns.barplot(x='State',y='Profit',data=data)
plt.xticks(rotation=90)
plt.show()
```

Here we coclude that some state have profit is in negative i.e loss .State such as florida,north carolina ,texas and many more

```python
plt.figure(figsize=(12,8))
sns.set_style('whitegrid')
sns.barplot(x='Category',y='Profit',data=data,hue='Sub-Category')
plt.xticks(rotation=90)
plt.show()
```

here we conclude that sub-category Table ,Bookcases are in loss

```python
plt.figure(figsize=(15,9))
sns.scatterplot(x='Profit',y='Sales',data=data,hue='Quantity',palette='rainbow')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,9))
sns.scatterplot(x='Profit',y='Sales',data=data,hue='Discount',palette='rainbow')
plt.xticks(rotation=90)
plt.show()
```

if there is high discount it causes loss
