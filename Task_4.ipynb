## library Imported

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
data= pd.read_csv('iris.csv')
data.set_index('Id',drop=True,inplace=True)
print('Data sucessfully imported')
data.head(10)
```

## Checking for missing value

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data_new=data.drop(['Species'],axis=1)
```

## Data Visualisation


```python
plt.figure(figsize=(12,8))
sns.boxplot(x=data['Species'],y=data['SepalLengthCm'],data=data)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x=data['Species'],y=data['SepalWidthCm'],data=data)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x=data['Species'],y=data['PetalWidthCm'],data=data)
plt.show()
```

## Data Training

```python
X=data_new
y=data['Species']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.metrics import confusion_matrix,accuracy_score,r2_score
dec=DecisionTreeClassifier()
dec.fit(X_train,y_train)
pred=dec.predict(X_test)
```

```python
print('accuracy_score: ',accuracy_score(y_test,pred))
print('confusion_matrix:\n',confusion_matrix(y_test,pred))
```

```python
plt.figure(figsize=(12,7))
sns.heatmap(confusion_matrix(y_test,pred),cmap='coolwarm',annot=True)
plt.show()
```

## How Decision Tree 


```python
coloum=list(data_new.columns)
plt.figure(figsize=(15,15))
plot_tree(dec.fit(X,y),feature_names=coloum,filled=True,proportion=True,rounded=True)
plt.show()
```
